```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('companies.csv')
```

```python
df.head()
```

```python
sns.countplot(y='state', data=df)

plt.show()
```

```python
job_openings = df[['Industry', 'job_openings']].groupby(['Industry']).sum().sort_values(by='job_openings', ascending=False).head(10)
```

```python
sns.barplot(x='job_openings', y=job_openings.index, data=job_openings)
```

```python
df[['founded', 'growth_percentage']].groupby(['founded']).mean().head(10)
```

```python
founded_average = df[['founded', 'growth_percentage']].groupby(['founded']).mean().sort_values('growth_percentage', ascending=True)
```

```python
sns.lineplot(x='founded', y='growth_percentage', data=founded_average)
```

```python
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
```

```python
df[['founded', 'company_name']].groupby(['founded']).value_counts()
```

```python
founded_year = df[['founded', 'company_name']].groupby(['founded']).value_counts().reset_index()
```

```python
founded_year = founded_year.rename(columns={0: 'company_count'})
```

```python
sns.histplot(x='founded', data=founded_year)
```
