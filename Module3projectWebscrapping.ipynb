```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup
```

```python
table = soup.find('table')
```

```python
if table:
    print('Table found')
else:
    print('Table not found')
```

```python
soup.find('table').find_all('tr', class_='book')[0].find_all('td')[1].text
```

```python
for employee in soup.find('table').find_all('tr', class_='book'):
  print(employee.find_all('td')[1].text)
```

```python
name=[]

for employee in soup.find('table').find_all('tr', class_='book'):
  name.append(employee.find_all('td')[1].text)
```

```python
name
```

```python
id=[]
name=[]
author=[]

for employee in soup.find('table').find_all('tr', class_='book'):
  id.append(employee.find_all('td')[0].text)
  name.append(employee.find_all('td')[1].text)
  author.append(employee.find_all('td')[2].text)
```

```python
id
```

```python
name
```

```python
author
```

```python
pd.DataFrame({'id':id, 'name':name, 'author':author})
```

```python
df=pd.DataFrame({'id':id, 'name':name, 'author':author})
```

```python
df.head()
```

```python
df['name'].groupby(df['author']).count().sort_values(ascending=False)
```

```python
df['author'].groupby(df['name']).count().sort_values(ascending=False)
```

```python
pages=[]
topic=[]
price=[]

for employee in soup.find('table').find_all('tr', class_='book'):
  pages.append(employee.find_all('td')[4].text)
  topics.append(employee.find_all('td')[5].text)
  price.append(employee.find_all('td')[6].text)
```

```python
pd.DataFrame({'pages':pages, 'topic':topics, 'price':price})
```

```python
df=pd.DataFrame({'pages':pages, 'topic':topics, 'price':price})
```

```python
df.head()
```

```python
df[['topic', 'price', 'pages']].sort_values(by=['topic', 'price'],ascending=[True, False])
```

```python
df['price'] = df['price'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
df['price'].groupby(df['topic']).mean().sort_values(ascending=False)
```

```python
df['pages'] = df['pages'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```
